```python
# For each row, if the value in 'country.etc' appears anywhere in 'name', remove it and any preceding space from 'name'
def clean_name(row):
    country = str(row['country.etc'])
    name = str(row['name'])
    # Look for the country code preceded by a space
    search_str = " " + country
    if country and search_str in name:
        # Remove the country code and preceding space
        return name.replace(search_str, "")
    elif country and country in name:
        # Remove the country code if present without preceding space
        return name.replace(country, "")
    else:
        return name

df['name'] = df.apply(clean_name, axis=1)
df.head()
```

- first name *
- last name *
- email *
- cell phone number *
- event address *
- event occasion *
- guest count
- catering budget
- event time *
- event date *

* indicates a required field 

Key takeaways: 
- One insight is that some emails may need some variety of these kinds of information. One key issue in treatment delivery could be in providng an event address. Wherever possible, it would be best to make it vague and possibly to orient the message to say that we'd be picking up the order rather than needing it to be delivered or asking for actual catering services. One key insight is that there's not really a way to select between catering that offers pickup vs catering that assumes serving the food themselves. One key issue might be getting responses saying they only do one or the other, and how to handle and validate that. 

- Additional, another assumption that's cropped up in our design is that some restaurants may require or expect that catering requests are exclusively submitted through their website.

```python
def decode_b64_string(s):
    # Pad with '=' to ensure the length is a multiple of 4
    padding = '=' * ((4 - len(s) % 4) % 4)
    return base64.b64decode(s + padding).decode('utf-8')

encoded_key = "YXV0aDB8NjhmNjgwMzgwZmFhYzg1NGU4MmZlN2NmfDFkNzE4ZjE0MzM"
try:
    decoded = decode_b64_string(encoded_key)
    print(f"Decoded API key: {decoded}")
except Exception as exc:
    print("Could not decode:", exc)
```

```python
import requests
import pandas as pd
import time
from email_validator import validate_email, EmailNotValidError
import dns.resolver

API_KEY = "YXV0aDB8NjhmNjgwMzgwZmFhYzg1NGU4MmZlN2NmfDFkNzE4ZjE0MzM"
API_URL = "https://api.targetron.cloud/data/places"

cities_df = pd.read_csv('sample_10-20-25.csv')

def clean_name(row):
    country = str(row['country.etc'])
    name = str(row['name'])
    search_str = " " + country
    if country and search_str in name:
        return name.replace(search_str, "")
    elif country and country in name:
        return name.replace(country, "")
    else:
        return name

cities_df['name'] = cities_df.apply(clean_name, axis=1)

union_cities = cities_df[cities_df['group'] == 'Union'].sample(n=124, random_state=42)
confederate_cities = cities_df[cities_df['group'] == 'Confederate'].sample(n=124, random_state=42)
selected_cities = pd.concat([union_cities, confederate_cities])

print(f"Selected {len(selected_cities)} cities:")
print(f"  Union: {len(union_cities)}")
print(f"  Confederate: {len(confederate_cities)}")

all_restaurants = []
api_calls = 0
max_budget = 497

for idx, row in selected_cities.iterrows():
    city = row['name']
    state = row['country.etc']
    group = row['group']
    
    if api_calls >= max_budget:
        print(f"\nReached budget limit ({max_budget} records)")
        break
    
    params = {
        'cc': 'US',
        'city': city,
        'state': state,
        'type': 'restaurant',
        'att': 'Dining options.Catering',
        'limit': 2
    }
    
    headers = {'X-API-KEY': API_KEY}
    
    try:
        response = requests.get(API_URL, params=params, headers=headers, timeout=30)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get('status') == 'Success':
                restaurants = data.get('data', [])
                
                for r in restaurants:
                    r['sampled_city'] = city
                    r['sampled_state'] = state
                    r['geographic_group'] = group
                    all_restaurants.append(r)
                    api_calls += 1
                
                print(f"{city}, {state} ({group}): {len(restaurants)} restaurants | Total: {api_calls}")
            else:
                print(f"{city}, {state}: {data.get('message', 'Unknown error')}")
        else:
            print(f"{city}, {state}: HTTP {response.status_code}")
    
    except Exception as e:
        print(f"{city}, {state}: {e}")
    
    time.sleep(0.5)
    
    if (idx + 1) % 25 == 0:
        print(f"\n--- Progress: {idx + 1}/{len(selected_cities)} cities, {api_calls} records ---\n")

print(f"\n{'='*80}")
print(f"COLLECTION COMPLETE")
print(f"{'='*80}")
print(f"Total restaurants collected: {len(all_restaurants)}")
print(f"API calls made: {api_calls}")
print(f"Estimated cost: ${api_calls * 0.02:.2f}")

df_raw = pd.DataFrame(all_restaurants)

df_raw.to_csv('targetron_pilot_raw.csv', index=False)
print(f"\nSaved raw data to: targetron_pilot_raw.csv")

EXCLUDE_CHAINS = [
    'mcdonald', 'subway', 'chipotle', 'panera', 'jimmy john',
    'dominos', 'pizza hut', 'taco bell', 'wendy', 'burger king',
    'arby', 'chick-fil-a', 'five guys', 'popeyes', 'kfc', 'panda express'
]

def is_chain(name):
    name_lower = str(name).lower()
    return any(chain in name_lower for chain in EXCLUDE_CHAINS)

df_filtered = df_raw[~df_raw['display_name'].apply(is_chain)].copy()

print(f"\nAfter filtering chains:")
print(f"  Raw: {len(df_raw)}")
print(f"  Filtered: {len(df_filtered)} ({len(df_filtered)/len(df_raw)*100:.1f}%)")

def validate_email_syntax(email):
    try:
        validate_email(email, check_deliverability=False)
        return True
    except:
        return False

def check_mx(email):
    try:
        domain = email.split('@')[1]
        dns.resolver.resolve(domain, 'MX')
        return True
    except:
        return False

if 'email_1' in df_filtered.columns:
    df_filtered['email_syntax_valid'] = df_filtered['email_1'].apply(
        lambda x: validate_email_syntax(x) if pd.notna(x) else False
    )
    df_filtered['email_mx_valid'] = df_filtered.apply(
        lambda row: check_mx(row['email_1']) if row['email_syntax_valid'] else False,
        axis=1
    )
    df_filtered['email_valid'] = df_filtered['email_syntax_valid'] & df_filtered['email_mx_valid']
    
    valid_emails = df_filtered['email_valid'].sum()
    print(f"\nEmail validation:")
    print(f"  Valid emails: {valid_emails} ({valid_emails/len(df_filtered)*100:.1f}%)")
else:
    print("\nNo email_1 field found in data")

df_filtered.to_csv('targetron_pilot_filtered.csv', index=False)
print(f"\nSaved filtered data to: targetron_pilot_filtered.csv")

print(f"\n{'='*80}")
print("GEOGRAPHIC DISTRIBUTION")
print(f"{'='*80}")
print(df_filtered.groupby('geographic_group').size())

if 'email_valid' in df_filtered.columns:
    print(f"\nVALID EMAILS BY GROUP:")
    print(df_filtered[df_filtered['email_valid']].groupby('geographic_group').size())

print(f"\n{'='*80}")
print("READY FOR PILOT DATA SUBMISSION")
print(f"{'='*80}")
```
